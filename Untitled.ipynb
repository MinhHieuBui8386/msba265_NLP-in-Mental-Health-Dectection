```python
import librosa
import os

# Path to the folder containing your audio files
audio_folder = r'C:\Users\nehal\OneDrive\Desktop\UOP\GradSchool\Fall2024\MSBA 265\Project\AudioWAV'

# Get a list of all .wav files in the folder
audio_files = [f for f in os.listdir(audio_folder) if f.endswith('.wav')]

# Iterate over each audio file
for audio_file in audio_files:
    audio_path = os.path.join(audio_folder, audio_file)
    
    # Check if the file exists (for safety)
    if os.path.exists(audio_path): 
        try:
            # Load the audio file using librosa
            y, sr = librosa.load(audio_path, sr=None)
            # Print the sample rate and duration of the audio file
            print(f"Loaded {audio_file} - Sample Rate: {sr} Hz, Duration: {len(y)/sr:.2f} seconds")
        except Exception as e:
            print(f"Error loading {audio_file}: {e}")
    else:
        print(f"File does not exist: {audio_path}")
```

```python
import librosa.display
import numpy as np

# Extract MFCCs from the audio file
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Print the shape of the MFCC array (time x coefficients)
print(f'MFCC Shape: {mfcc.shape}')

# Visualize the MFCCs
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfcc, x_axis='time', sr=sr)
plt.colorbar()
plt.title(f'MFCCs for {audio_file}')
plt.show()
```

```python
# Flatten the MFCCs into a 1D array for machine learning models
mfcc_flattened = mfcc.flatten()

# Print the flattened MFCC shape
print(f'Flattened MFCC shape: {mfcc_flattened.shape}')
```

```python
import numpy as np
import librosa
import librosa.display
import os

# Initialize empty lists to store features and labels
features = []
labels = []

# Define the fixed length for MFCC (e.g., 13 MFCCs over 130 time frames)
fixed_length = 130  # You can adjust this based on your dataset

# Loop through all audio files
for audio_file in audio_files:
    audio_path = os.path.join(audio_folder, audio_file)
    
    if os.path.exists(audio_path): 
        try:
            # Load the audio file
            y, sr = librosa.load(audio_path, sr=None)

            # Extract MFCCs (13 MFCCs are commonly used)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

            # Pad or truncate MFCC to the fixed length
            if mfcc.shape[1] < fixed_length:
                # Pad with zeros if the length is shorter
                mfcc = np.pad(mfcc, ((0, 0), (0, fixed_length - mfcc.shape[1])), mode='constant')
            else:
                # Truncate if the length is longer
                mfcc = mfcc[:, :fixed_length]

            # Flatten MFCCs for use in machine learning models
            mfcc_flattened = mfcc.flatten()

            # Add the features and labels to the respective lists
            features.append(mfcc_flattened)
            
            # Example: Assuming labels are derived from file names (adjust accordingly)
            label = audio_file.split('_')[0]  # Replace with your own labeling strategy
            labels.append(label)

        except Exception as e:
            print(f"Error loading {audio_file}: {e}")
    else:
        print(f"File does not exist: {audio_path}")

# Convert the feature list into a numpy array
X = np.array(features)
y = np.array(labels)

# Check the shape of the features array
print(f"Shape of X: {X.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features (optional but recommended)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train an SVM classifier
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```
